```python
import numpy as np
import matplotlib.pyplot as plt
width = 2.5 #By default, Matplotlib creates figures with a width of 6.4 inches and a height of 4.8 inches
height = 2.5
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
fig, ax = plt.subplots(figsize=(2.5, 2.5))
ax.plot(x,y)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(7.5, 2.5))

x = np.linspace(0, 10, 100)

y1 = np.cos(x)
ax1.set_facecolor("#ddddff")
ax1.plot(x,y1, color="black")

y2 = np.sin(x)
ax2.spines["top"].set_color("None")
ax2.spines["right"].set_color("None")
ax2.plot(x,y2, linewidth=3)

y3 = np.tan(x)
y4 = np.sin(x)
ax3.set_facecolor("#ffffdf")
ax3.plot(x, y3, x, y4, linestyle="--")

plt.plot()
```

```python
plt.figure(figsize=(width, height))
x = np.linspace(1, 100, 20)
y1 = np.random.random(20)
y2 = np.random.random(20)

plt.scatter(x, y1, color='blue', label='Dataset 1')
plt.scatter(x, y2, color='green', label='Dataset 2')

plt.legend()
plt.show()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(width*4, height))
x = np.linspace(1, 100, 20)
y1 = np.random.random(20)
y2 = np.random.random(20)
y3 = np.random.random(20)
y4 = np.random.random(20)

ax1.scatter(x, y1, marker='o') #scatter_markers = [".", ",", "o", "v", "^", "<", ">", "1", "2", "s", "p", "*", "h", "H", "+", "x", "|", "_", "P", "X"]
ax2.scatter(x, y2, c='red', s=8)

z3 = np.random.randint(0, 10, size=20)
ax3.scatter(x, y3, c=z3, cmap='viridis', marker='*')

z4 = np.concatenate((np.zeros(10),np.random.random(5),  np.ones(5)))
ax4.scatter(x, y4, c=z4, cmap='viridis', marker="X")

plt.show()
```

```python
plt.figure(figsize=(width, height))
y = np.random.random(100)
plt.hist(y)
plt.show()
```

```python
plt.figure(figsize=(width, height))
z = np.random.uniform(0, 4, 4)
plt.pie(z)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
z1 = np.zeros((8, 8))
z2 = np.random.uniform(0, 1, (8, 8))
z3 = np.random.uniform(0, 1, (8, 8))
z4 = np.ones((8, 8))

ax[0, 0].imshow(z1)
ax[0, 0].set_title("zx1")

ax[0, 1].imshow(z2)
ax[0, 1].set_ylabel("y_axis")
ax[0, 1].set_xlabel("x_axis")

ax[1, 0].imshow(z3)
ax[1, 1].imshow(z4)

plt.show()
```
